```python

London_Borough_U_List=[]

def Borough_Pricing(London_Borough='All',Base_Year=1998, Comp_Year=2018,Counter=0):
    """Function is designed to calculate the annual mean price for either a Borough or all Boroughs."""
    
    London_Borough_U_List=prop_detail['London_Borough'].unique().tolist()
    
    try:
            
        if London_Borough=='All':
            

            Borough_Counter= len(London_Borough_U_List)

            while Counter < Borough_Counter:

                Base_Year_AVG_1= prop_detail.loc[(prop_detail['London_Borough'] == London_Borough_U_List[Counter])&\
                                                       (prop_detail['Year'] == Base_Year ),: 'Average Price'].mean()

                Comp_Year_AVG_1= prop_detail.loc[(prop_detail['London_Borough'] == London_Borough_U_List[Counter]) &\
                                                       (prop_detail['Year'] == Comp_Year ),: 'Average Price'].mean()

                Price_Change = Comp_Year_AVG_1 - Base_Year_AVG_1

                Ratio = Base_Year_AVG_1/Comp_Year_AVG_1

                print(f"Borough: {London_Borough_U_List[Counter]}\n{Base_Year} average price: {Base_Year_AVG_1[0]:,.2f}\
                \n{Comp_Year} average price: {Comp_Year_AVG_1[0]:,.2f}\nPrice movement: {Price_Change[0]:,.2f}\
                \nPrice ratio: {Ratio[0]:.2%}\n\n")
                Counter = Counter+1
        
        elif London_Borough not in London_Borough_U_List:
            print("Borough does not exists in dataset")
            return
        
        
        else:

            Base_Year_AVG_1= prop_detail.loc[(prop_detail['London_Borough'] == London_Borough)&\
                                               (prop_detail['Year'] == Base_Year ),: 'Average Price'].mean()

            Comp_Year_AVG_1= prop_detail.loc[(prop_detail['London_Borough'] == London_Borough) &\
                                               (prop_detail['Year'] == Comp_Year ),: 'Average Price'].mean()

            Price_Change = Comp_Year_AVG_1 - Base_Year_AVG_1

            Ratio = Base_Year_AVG_1/Comp_Year_AVG_1

        print(f"Borough: {London_Borough}\n{Base_Year} average price: {Base_Year_AVG_1[0]:,.2f}\
        \n{Comp_Year} average price: {Comp_Year_AVG_1[0]:,.2f}\nPrice movement: {Price_Change[0]:,.2f}\
        \nPrice ratio: {Ratio[0]:.2%}\n\n")

    except (ValueError):
        print("error")
```
